## 2. Muestreo sistemático

```python
def systematic_sampling(df, step):
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
```

```python
systematic_sample = systematic_sampling(econdata, 3)
systematic_sample.head()
```

## 3. Método estratificado

```python
econdata["stratified"] = econdata["delegacion"] + "," + econdata["tipo"]
(econdata["stratified"].value_counts()/len(econdata)).sort_values(ascending=False)
```

```python
def stratified_sampling(df, var, size):
    stratified_sample = pd.DataFrame()

    props = df[var].value_counts()/len(df)
    strata = props.index

    for s in strata:
        strat_size = size*props[s]
        stratum = df[df[var] == s]
        sample = stratum.sample(int(strat_size))
        stratified_sample = pd.concat([stratified_sample, sample])
    return stratified_sample

stratified_sample = stratified_sampling(econdata, "stratified", 90)
(stratified_sample["stratified"].value_counts()/len(stratified_sample)).sort_values(ascending=False)
```

```python
strat_values = ["Cuautémoc,Hotel", "Cuautémoc,Museo", "Venustiano Carranza,Hotel", "Cuauhtémoc,Mercado", "Venustiano Carranza,Mercado"]
strat_prop = [0.5, 0.2, 0.1, 0.1, 0.1]
strat_df = stratified_sampling(econdata, dict(zip(strat_values, strat_prop)), strat_values, "stratified", seed=42)
strat_df["stratified"].value_counts()/len(strat_df)
```
