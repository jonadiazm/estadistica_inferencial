```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Datos originales
data = np.array([12, 15, 14, 10, 13, 17, 20, 18, 16, 14])

# Parámetros de bootstrapping
n_iterations = 1000  # Número de muestras bootstrap
sample_size = len(data)  # Tamaño de cada muestra bootstrap
bootstrap_means = []  # Almacenará las medias bootstrap

# Bootstrapping
np.random.seed(42)
for i in range(n_iterations):
    sample = np.random.choice(data, size=sample_size, replace=True)  # Muestra con reemplazo
    bootstrap_means.append(sample.mean())

# Intervalo de confianza al 95%
lower_bound = np.percentile(bootstrap_means, 2.5)
upper_bound = np.percentile(bootstrap_means, 97.5)

print(f"Intervalo de confianza del 95% para la media: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
plt.hist(sample)
plt.xlabel('Media')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(bootstrap_means, bins=30, edgecolor='black')
plt.axvline(lower_bound, color='red', linestyle='--', label='2.5%')
plt.axvline(upper_bound, color='red', linestyle='--', label='97.5%')
plt.xlabel('Media')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
